Plot the distributions of number of deaths by each illness category

```python
"""
This script processes a DataFrame containing death statistics for Utah, filters it to include only rows where the measure is 'Deaths',
and then groups the data by cause of death and year to calculate the mean values. It then creates a 2x2 grid of subplots to visualize
the number of deaths over the years for different causes, including confidence intervals.
Variables:
    death_df (DataFrame): Filtered DataFrame containing only rows where the measure is 'Deaths'.
    causes (ndarray): Array of unique cause names from the DataFrame.
    group_cols (list): List of columns to be used for grouping the data.
    death_group_cause (DataFrame): Grouped DataFrame with mean values of deaths by cause and year.
    fig (Figure): Matplotlib figure object for the subplots.
    ax (ndarray): Array of axes objects for the subplots.
    colors (list): List of colors to be used for plotting different causes.
Plot:
    A 2x2 grid of subplots, each showing the number of deaths over the years for a specific cause, with confidence intervals.
"""


death_df = utah_df_num.loc[utah_df['measure_name'] == 'Deaths']
death_df = death_df.drop(columns=['measure_name'])

causes = utah_df_num['cause_name'].unique()
group_cols = ['cause_name', 'year', 'val', 'upper', 'lower']
death_group_cause = death_df[group_cols].groupby(['cause_name', 'year']).mean()
fig, ax = plt.subplots(figsize=(20, 10), nrows=2, ncols=2)
ax = ax.flatten()

colors = ['b', 'g', 'r', 'y']
for i, (cause, group) in enumerate(death_group_cause.groupby(level=0)):
    years = group.index.get_level_values('year')
    ax[i].plot(years, group['val'], label="Death", color=colors[i])
    ax[i].fill_between(years, group['upper'], group['lower'], color=colors[i], alpha=0.2, label='Confidence Interval')
    ax[i].set_title(f'{cause}')
    ax[i].set_xlabel('Year')
    ax[i].set_ylabel('Deaths')
    ax[i].legend()

fig.suptitle('Deaths by Causes', fontsize=15)
plt.legend()
plt.show()
    
```

### Tranform Data for the Regression Model

One-hot encode rei_name column to handle categorical type

Groups the death_df DataFrame by `year`, `cause_name`, `Air pollution`, and `Particulate matter pollution` columns,
and calculates the mean of the numeric columns for each group.

```python
"""
Groups the death_df DataFrame by 'year', 'cause_name', 'Air pollution', and 'Particulate matter pollution' columns,
and calculates the mean of the numeric columns for each group.

Returns:
    DataFrame: A DataFrame with the mean values of the numeric columns for each group, with the grouping columns reset to regular columns.
"""

grouped = death_df.groupby(
    ['year', 'cause_name','Air pollution', 'Particulate matter pollution']) \
    .mean(numeric_only=True).reset_index()
grouped
```

Pivot the data such that there's a column for each type of respiratory illnesses. Each year will have 2 rows, adjusting for `Ait polllution` and `Particulate matter pollution`. Later, we can fit time-series data into regression model, where each illness column would be a depedent variable.

```python
pivoted = death_df.pivot_table(
    index=['year', 'Air pollution', 'Particulate matter pollution'],
    columns='cause_name',
    values=['val']
)


# Flatten the MultiIndex columns
pivoted.columns = [col[1] for col in pivoted.columns.values]
pivoted.reset_index(inplace=True)
pivoted.head(10)
```

### Read Intermediate data from Part 1


```python
nearby_fires_df = pd.read_csv('../data/intermediate/nearby_fires.csv')
cumulative_smoke_df = pd.read_csv('../data/intermediate/cumulative_smoke.csv')
gas_aqi_data = pd.read_csv('../data/intermediate/gas_df_filtered.csv')
part_aqi_data = pd.read_csv('../data/intermediate/part_df_filtered.csv')


display(nearby_fires_df.head(5))
display(cumulative_smoke_df.head(5))
display(gas_aqi_data.head(5))
display(part_aqi_data.head(5))
```

### Transform Annual AQI & Fire Data for Time-series Model

Aggrgate AQI, Fire, and Smoke data such that each year will have 1 row by finding the average of year. Pivot the data such that there's a column for each metric. Later, we can fit these time-series data into regression model such that each column is a predictor.

```python
aqi_cols = ['parameter', 'aqi', 'year']

print(gas_aqi_data['parameter'].unique())
print(part_aqi_data['parameter'].unique())
gas_aqi_annual = gas_aqi_data[aqi_cols].groupby(['parameter', 'year']).mean().reset_index()
part_aqi_annual = part_aqi_data[aqi_cols].groupby(['parameter', 'year']).mean().reset_index()


# Pivot the Gas & Part Data
gas_pivot = gas_aqi_annual.pivot(index='year', columns='parameter', values='aqi')
gas_pivot.columns.name = None  # Remove the name for columns
gas_pivot.reset_index(inplace=True)  # Bring 'year' back as a column

part_pivot = part_aqi_annual.pivot(index='year', columns='parameter', values='aqi')
part_pivot.columns.name = None  # Remove the name for columns
part_pivot.reset_index(inplace=True)  # Bring 'year' back as a column

display(gas_pivot.head(5))
display(part_pivot.head(5))
```

```python
annual_fire = nearby_fires_df[['Fire_Year','GIS_Acres','Distance']].groupby('Fire_Year').mean().reset_index()
annual_fire['Fire_Year'] = annual_fire['Fire_Year'].astype(int)
annual_fire.head(5)
```

### Combine all annual data

```python
# Stack the Gas & Part Data
predictors_df = pd.merge(gas_pivot, part_pivot, on='year', how='outer')
predictors_df = pd.merge(annual_fire, predictors_df, left_on='Fire_Year', right_on='year', how='outer')
predictors_df = pd.merge(cumulative_smoke_df, predictors_df, on='year', how='outer')
predictors_df = predictors_df.drop(columns=['Fire_Year'])

predictors_df.head(15)
```

Sometimes, the joins don't behave as expect, so we double-check that there're no rows where year is NaN.

```python
null_values = predictors_df.isnull().sum()
print(null_values)

cleaned_df = predictors_df.dropna(subset=['year'])
cleaned_df
```

### Predictive Model

#### Combine Annual Health Data with Annual Intermediary Data

After combining all data, there are some cells with null values. So we fill those using backward-fill and then forward-fill. This is because I suspect there's increasing or decreasing trends.

```python
# Combine with pivoted table from the previous step
df_combined = pd.merge(pivoted, cleaned_df, on='year', how='left')
df_combined

null_values = df_combined.isnull().sum()
print(null_values)

# Fill missing values
final_df = df_combined.fillna(method='bfill', axis=0)  # Backfill along the rows
final_df = df_combined.fillna(method='ffill', axis=0)  # Forward fill along the rows

final_df
```

#### Fit a Linear Regression

Now that the data is ready, we will fit a multivariate Linear Regression to see which predictors consistenly contribute to the number of deaths by illnesses. 

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm

# Define the predictor variables (X) and target variables (Y)
# 'Air pollution', 'Particulate matter pollution', 
X_columns = ['cumulative_smoke', 'GIS_Acres', 'Distance',
             'Carbon monoxide', 'Nitrogen dioxide (NO2)', 'Ozone', 'Sulfur dioxide',
             'Acceptable PM2.5 AQI & Speciation Mass', 'PM10 Total 0-10um STP', 'PM2.5 - Local Conditions']
Y_columns = ['Lower respiratory infections', 'Otitis media', 
             'Respiratory infections and tuberculosis', 'Upper respiratory infections']

# Split data into predictors (X) and targets (Y)
X = final_df[X_columns].fillna(0)  # Fill missing values as needed
Y = final_df[Y_columns].fillna(0)
# X.loc[:,['Air pollution', 'Particulate matter pollution']] = X[['Air pollution', 'Particulate matter pollution']].astype(int)

# Adding a constant for the intercept
X = sm.add_constant(X)

# Fit models for each target (illness) separately
results = {}
for illness in Y_columns:
    # Initialize and fit the model for the current illness
    model = sm.OLS(np.asarray(Y[illness]), X).fit()
    
    # Store results
    results[illness] = {
        'model': model,
        'summary': model.summary()
    }
    
    # Print metrics and p-values
    print(f"Results for {illness}:")
    print(f"R^2 = {model.rsquared:.4f}")
    print(f"MSE = {mean_squared_error(Y[illness], model.predict(X)):.4f}\n")
    print(model.summary())

# Example: Access coefficients and p-values for one illness
for illness in Y_columns:
    coefficients = results[illness]['model'].params
    p_values = results[illness]['model'].pvalues
    significant = p_values < 0.05  # Identify significant variables (p < 0.05)

    print(f"\nSignificant predictors for {illness}:")
    print(coefficients[significant])
    # print(f"Summary: {results[illness]['summary']}")
```

Plot Actual vs. Predicted Values


```python
Y_pred = np.array([model.predict(X) for model in results.values()]).
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
axes = axes.flatten()

for i, illness in enumerate(Y_columns):
    axes[i].plot(Y[illness], label='Actual', color='blue')
    axes[i].plot(Y_pred[:, i], label='Predicted', color='red', linestyle='--')
    axes[i].set_title(f'Actual vs. Predicted for {illness}', fontsize=14)
    axes[i].set_xlabel('Index', fontsize=12)
    axes[i].set_ylabel('Value', fontsize=12)
    axes[i].legend()
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

Plot significant predictors from the model to see their relative relationship with the dependent variables

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 16))
axes = axes.flatten()
for i, illness in enumerate(Y_columns):
    axes[i] = plt.subplot(2, 2, i + 1)
    coefficients = results[illness]['model'].params
    p_values = results[illness]['model'].pvalues
    significant = p_values[(p_values < 0.05) & (p_values.index != 'const')].index
    significant_coefficients = coefficients[significant]

    # Sort coefficients by magnitude
    sorted_coefficients = significant_coefficients.sort_values(key=abs, ascending=False)

    # Color coding: green for positive, red for negative relationships
    colors = ['green' if val > 0 else 'red' for val in sorted_coefficients]

    # Plot bar chart
    sorted_coefficients.plot(kind='bar', color=colors, edgecolor='black')

    # Add titles and labels
    plt.title(f"{illness}", fontsize=20)
    plt.xlabel("Predictors")
    plt.ylabel("Coefficient Value")
    plt.xticks(rotation=10, ha='right', fontsize=13)
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

    # Show the plot
plt.suptitle('Significant Predictors for Number of Deaths by Respiratory Illness', fontsize=20)
plt.tight_layout()

plt.show()


```
